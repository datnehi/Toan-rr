```python
# 2) Trong các tập con của tập {1,2,...,9} xếp theo thứ tự từ điển, tìm tập con đứng ngay sau tập {1, 2, 5, 6}
def next_combinations(n, a):
  r = len(a)
  i = r - 1
  while i >= 0 and a[i] == n - r + (i + 1):
    i -= 1
  if i == -1:
    return None

  return a[:i] + [a[i] + j for j in range(1, r-i+1)]

next_combinations(9, [1, 2, 5, 6, 7, 9])
```

```python
# 3) Tìm hoán vị đứng sau hoán vị 265431 theo thứ tự từ điển
def next_permutations(a):
  n = len(a)
  k = n - 1
  while k >= 1 and a[k - 1] > a[k]:
    k -= 1

  if k == 0:
    return None

  i = n - 1
  while a[i] < a[k - 1]:
    i -= 1
  a[k - 1], a[i] = a[i], a[k - 1]                                                                  
  b = a[k:]
  b.reverse()
  return a[:k] + b

next_permutations([1, 6, 5, 4, 3, 2])
```

```python
# 18) Có bao nhiêu cách sắp xếp 2 bi xanh, 2 bi đỏ, 3 bi vàng
def permutations_with_replacement(a, n):
  r = len(a)
  if sum(n) == 0:
    return [[]]
  P = []
  for i in range(r):
    if n[i] > 0:
      n_ = n.copy()
      n_[i] -= 1
      for p in permutations_with_replacement(a, n_):
        p = [a[i]] + p
        P.append(p)
  return P
len(permutations_with_replacement(['X', 'Đ', 'V'] , [3 , 2 , 4]))
```

```python
#19) Trong các hoán vị của 4 chữ a, 3 chữ b, và 3 chữ c, tìm hoán vị thứ 15 theo thứ tự từ điển (viết hoán vị dưới dạng xâu)
def permutations_with_replacement(a, n):
  r = len(a)
  if sum(n) == 0:
    return [[]]
  P = []
  for i in range(r):
    if n[i] > 0:
      n_ = n.copy()
      n_[i] -= 1
      for p in permutations_with_replacement(a, n_):
        p = [a[i]] + p
        P.append(p)
  return P

permutations_with_replacement(['a', 'b', 'c'], [4, 3, 3])
# Kết quả: ['a', 'a', 'a', 'a', 'c', 'b', 'c', 'b', 'c', 'b']
```

```python
# Trong các tổng riêng của n = 7, xếp theo thứ tự từ điển, tìm tổng riêng thứ 20(viết dưới dạng xâu)
def summands(n):
    if n == 1:
        return [[1]]
    S = []
    for i in range(1, n):
        for s in summands(n-i):
            s = [i] + s
            S.append(s)
    S.append([n])
    return S

def find_20th_summand(n):
    all_summands = summands(n)
    # Sắp xếp danh sách all_summands theo thứ tự từ điển
    sorted_summands = sorted(all_summands)
    # Trả về tổng riêng thứ 20
    return sorted_summands[225]  # Vị trí 20 - 1 vì list đếm từ 0

# Tìm tổng riêng thứ 20 của n = 7
result = find_20th_summand(9)
print(result)
```
